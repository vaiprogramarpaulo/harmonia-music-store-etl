```python

# Harmonia Music Store Project Using ETL Pipeline
# LinkedIn: paulovrsouza

import pandas as pd

# ============================
# Extract
# ============================

def extract(path: str) -> pd.DataFrame:
    # Etapa de EXTRAÇÃO: 
    # Aqui, carregamos os dados brutos a partir de um arquivo CSV.
    
    # O resultado é um DataFrame do Pandas que será a base das próximas etapas.
    return pd.read_csv(path, sep=";")

# ============================
# Transform
# ============================

def transform(df: pd.DataFrame) -> pd.DataFrame:
    # Etapa de TRANSFORMAÇÃO:
    # Nesta fase, limpamos e organizamos os dados para facilitar a análise.

    # 1. Removemos linhas completamente nulas e duplicadas
    df = df.dropna(how='all')
    df = df.drop_duplicates()

    # 2. Criamos colunas derivadas que ajudam na análise
    # Abaixo, convertemos meses em números para ordenar corretamente:
    ordem_meses = { "Janeiro": 1, "Fevereiro": 2, "Março": 3, "Abril": 4,
    "Maio": 5, "Junho": 6, "Julho": 7, "Agosto": 8,
    "Setembro": 9, "Outubro": 10, "Novembro": 11, "Dezembro": 12
    }
    df["Mês_Num"] = df["Mês"].map(ordem_meses)

    # Abaixo, calculamos o valor do Ticket Médio (Receita Total / Quantidade):
    df["Ticket Médio"] = df["Receita Total"] / df["Quantidade"]

    # 3. Fazemos um ajuste estético: índice começa em 1 para facilitar leitura
    df.index = df.index + 1

    return df
    
# ============================
# Load
# ============================

def load(df: pd.DataFrame, path_csv: str, path_excel: str) -> None:
    # Etapa de CARREGAMENTO:
    # Por último, exportamos os dados já tratados para novos arquivos.
    # Dessa forma, podemos usá-los em outras ferramentas, como Power BI
    
    # Salvamos em CSV e Excel, respectivamente, com index=False para não incluir o índice como coluna extra
    df.to_csv(path_csv, sep=";", index=False) 
    df.to_excel(path_excel, index=False)

# ============================
# Pipeline Principal
# ============================

def pipeline():
    # O pipeline organiza todas as etapas de forma sequencial:

    # 1. Extrai os dados brutos do arquivo em CSV informado entre aspas
    df = extract("harmonia_music_store.csv")
    
    # 2. Transforma os dados (limpeza e criação de colunas úteis)
    df = transform(df)

    # 3. Carrega os dados tratados em novos arquivos
    load(df, "harmonia_music_store_tratado.csv", "harmonia_music_store_tratado.xlsx")

    # 4. Mostra uma prévia dos dados tratados, ou do DataFrame, para validação
    print(df.head())

# Executa o pipeline
pipeline()
```
